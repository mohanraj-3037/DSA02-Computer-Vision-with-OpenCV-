```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img=cv2.imread('inputa.jpg',0)
plt.imshow(img,cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

```python
img1=cv2.imread('inputb.jpg')
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
blur=cv2.GaussianBlur(img1,(15,15),0)
plt.imshow(blur)
plt.title("Gaussian Blur Image")
plt.axis("off")
plt.show()
```

```python
img2=cv2.imread('inputc.jpg')
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
canny=cv2.Canny(img2,100,200)
plt.imshow(canny,cmap='gray')
plt.title("Canny Edge Image")
plt.axis("off")
plt.show()
```

```python
img3 = cv2.imread('inputd.jpg', 0)

canny = cv2.Canny(img3, 100, 200)

kernel = np.ones((5,5), np.uint8)

dilate = cv2.dilate(canny, kernel, iterations=1)

plt.imshow(dilate, cmap='gray')
plt.title("Dilated Image")
plt.axis("off")
plt.show()
```

```python
img4=cv2.imread('inpute.jpg',0)
canny=cv2.Canny(img4,100,200)
kernel=np.ones((5,5),np.uint8)
dilate=cv2.dilate(canny,kernel,iterations=1)
erode=cv2.erode(dilate,kernel,iterations=1)
plt.imshow(erode,cmap='gray')
plt.title("Eroded Image")
plt.axis("off")
plt.show()
```
