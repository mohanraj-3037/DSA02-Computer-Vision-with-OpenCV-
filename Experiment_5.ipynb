```python
import cv2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("input.jpg")
cw=cv2.rotate(img,cv2.ROTATE_90_CLOCKWISE)
cww=cv2.rotate(img,cv2.ROTATE_90_COUNTERCLOCKWISE)
cv2.imwrite("cw.jpg",cw)
cv2.imwrite("cww.jpg",cww)
```
