```python
import cv2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("input.jpg")
h,w=img.shape[:2]
```

```python
src=np.float32([[0,0] ,[w-1,0],[0,h-1]])
aff=np.float32([[50,50],[w-100,30],[70,h-80]])
M=cv2.getAffineTransform(src,aff)
out=cv2.warpAffine(img,M,(w,h))
cv2.imwrite("AffineT.img",out)
plt.imshow(out,[: ,: ,::-1]
plt.axis("off")
```
